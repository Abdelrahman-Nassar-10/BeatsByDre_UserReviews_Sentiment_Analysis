### 4.2 Data Collection
As mentioned the collected data was from reviews on amazon of the products using Oxylabs API using functions to fetch, convert and save the collected data to a CSV file, we used 100 reviews for each product to have an even distribution for our data netting us 1000 samples.

*follow the code below for more instructions on how the process was done:-*

```python
import pandas as pd
import google.generativeai as genai
import os
import json
import requests
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from textblob import TextBlob
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

```

```python


# the functions that serves the fetching part using oxylabs API.
def fetch_reviews_from_oxylabs(asin, username, password, page_number=1):
    payload = {
        'source': 'amazon_reviews',
        'user_agent_type': 'desktop',
        'parse': True,
        'domain': 'com',
        'pages': '20',
        'query': asin,
        'start_page': str(page_number)
    }
    response = requests.post(
        'https://realtime.oxylabs.io/v1/queries',
        auth=(username, password),
        json=payload
    )
    return response.json()



# the function that takes the json file and extracts the data from it into a dataframe.
def extract_reviews_from_json(json_data):
    reviews_data = []    # list that saves all the reviews
    try:                                              # try block to check if we were able to get the review details below
        for result in json_data.get('results', []):
            asin = result['content'].get("asin", "unknown_asin")
            for review in result['content'].get('reviews', []):
                review_info = {
                    'review_id': review.get('id', 'unknown_id'),
                    'product_id': asin,
                    "title": review.get("title", "no_title"),
                    "author": review.get("author", "anonymous"),
                    "rating": review.get("rating", "no_rating"),
                    "content": review.get("content", "no_content"),
                    "timestamp": review.get("timestamp", "no_timestamp"),
                    "profile_id": review.get("profile_id", "no_profile_id"),
                    "is_verified": review.get("is_verified", False),
                    "helpful_count": review.get('helpful_count', 0),
                    "product_attributes": review.get("product_attributes", "no_attributes")
                }
                reviews_data.append(review_info)
    except Exception as e:                       # if there's an error let us know
        print(f"Error extracting reviews: {e}")
    return reviews_data                          # return the reviews data



#function that saves the dataframes into a csv file.
def save_reviews_to_csv(reviews_data, output_csv_file_name):
    reviews_df = pd.DataFrame(reviews_data)       # turn list to dataframe first
    reviews_df.to_csv(output_csv_file_name, index=False)    # then turn to csv
    print(f"Reviews successfully saved to {output_csv_file_name}")
```

```python
## example usage.
save_reviews_to_csv(extract_reviews_from_json(fetch_reviews_from_oxylabs("B08WM3LMJF", "Nassar_87H4e", "Abdo_2002_Abdo", page_number=1)), "JBL_data.csv")
```

## 6. Sentiment Analysis

### 6.1 Methodology
**Here we conduct the sentiment analysis part to fully look at the whole picture in detail and capture the voice of the customers to understand them and use this feedback to find issues, give insights about certain things and provide improvments.**

and for sentiment analysis we used 2 analyzers **Textblob** and **VADER**:-


```python
#Sentiment Analysis using textblob
def sentiment_analysis(review):
    text=TextBlob(review)
    return{
        "Polarity":text.sentiment.polarity,
        "Subjectivity":text.sentiment.subjectivity
        }


#Sentiment Analysis using VADER
nltk.download('vader_lexicon')
analyzer=SentimentIntensityAnalyzer()
def vader_sent_anal(reviews):
    return analyzer.polarity_scores(reviews)
```

**Applying the VADER analyzer to the data.**

```python
sentiment_scores=review_df["content"].apply(vader_sent_anal)
sentiment_scores
```

**Just concatonating the analysis values to our dataframe.**

```python
sentimentdf=pd.json_normalize(sentiment_scores)
new_reviewdf=pd.concat([review_df,sentimentdf], axis=1)
```

**like the VADER here we apply and concatonate the textblob analysis to our dataframe.**

```python
textblobanalysis=review_df["content"].apply(sentiment_analysis)
textblobandf=pd.json_normalize(textblobanalysis)
new_reviewdf=pd.concat([new_reviewdf,textblobandf], axis=1)
```

### 8.2 SWOT Analysis
# Strengths
**Sound Quality**: The Beats Studio Pro is praised for its balanced audio profile, with crisp highs and deep bass. This positions it well for audiophiles seeking quality sound at a lower price point compared to premium competitors.

**Comfort and Design**: The headphones are noted for their comfortable fit, lightweight design, and good build quality. These aspects contribute to a positive user experience, especially for long-term wear.

**Apple Ecosystem Integration**: Seamless pairing, automatic device switching, and Siri activation provide a strong advantage for users within the Apple ecosystem, enhancing convenience and connectivity.

**Value for Money**: When purchased on sale, the Studio Pro offers high-quality features at an affordable price, making it a strong value proposition.

# Weaknesses
**Sound Limitations**: improvements in Transparency Mode and ANC effectiveness are needed. An adjustable EQ could address these concerns.

**Microphone Quality**: The microphone is considered average, with reported issues on platforms like Windows 11. Upgrading mic quality could improve user satisfaction.

**Design and Usability**: The need for a more durable headband, a better-designed case, and enhanced control usability highlights areas for improvement. Additionally, the absence of an auto-pause feature could be a drawback.

# Opportunities
**Product Enhancements**: Implementing improvements such as an adjustable EQ, better ANC, enhanced Transparency Mode, and upgraded mic quality can address user concerns and increase appeal.

**Design Upgrades**: Reinforcing the headband, improving the case design, and adding features like auto-pause can enhance the overall user experience and attract more customers.

**Expanding Market**: By addressing comfort and usability issues, Beats can better target remote workers, travelers, and gym enthusiasts, broadening its customer base.

**Improved Connectivity**: Resolving Bluetooth issues and simplifying instructions can enhance the user experience and reduce potential frustrations, fostering greater customer satisfaction.

# Threats
**Intense Competition**: The headphone market is highly competitive, with strong contenders like AirPods Max and other high-end models offering advanced features and superior performance.

**User Expectations**: As technology and user expectations evolve, failing to keep up with advancements in sound quality, noise cancellation, and user-friendly features could impact market position.

**Price Sensitivity**: While Beats Studio Pro offers good value, price-conscious consumers may still prefer lower-cost alternatives if perceived value isn’t high enough, especially if competitors offer similar features at lower prices.

**Brand Perception**: Any ongoing issues with comfort, mic quality, or other features could impact brand perception and lead to negative reviews, affecting overall sales and market reputation.
